```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno # For visualizing missing values
import gc
import os # For creating directories and joining paths

# --- Configuration ---
BASE_PATH = '/kaggle/input/amini-soil-prediction-challenge-dataset/' # Adjust if your files are in a subdirectory
PLOTS_DIR = 'eda_plots' # Directory to save plots
# Corrected list of satellite files based on the image provided
SATELLITE_FILES_INFO = {
    'LANDSAT8_data_updated.csv': 'ls8',
    'MODIS_MCD43A4_data.csv': 'mcd43',
    'MODIS_MOD09GA_data.csv': 'mod09',
    'MODIS_MOD11A1_data.csv': 'mod11a1',
    'MODIS_MOD13Q1_data.csv': 'mod13',
    'MODIS_MOD16A2_data.csv': 'mod16',
    'Sentinel1_data.csv': 's1',
    'Sentinel2_data.csv': 's2'
}
TARGET_COLUMNS = ['N', 'P', 'K', 'Ca', 'Mg', 'S', 'Fe', 'Mn', 'Zn', 'Cu', 'B']
SEED = 42

# --- Create plots directory if it doesn't exist ---
if not os.path.exists(PLOTS_DIR):
    os.makedirs(PLOTS_DIR)
print(f"Plots will be saved to: {os.path.abspath(PLOTS_DIR)}")

# --- Helper Function for Basic DataFrame Inspection ---
def inspect_dataframe(df, df_name="DataFrame"):
    print(f"\n--- Inspection for {df_name} ---")
    print(f"\nShape: {df.shape}")
    print(f"\nFirst 5 rows:\n{df.head()}")
    print(f"\nInfo:")
    df.info()
    print(f"\nBasic Statistics (Numeric Columns):\n{df.describe().T}")
    if not df.select_dtypes(include='object').empty:
        print(f"\nBasic Statistics (Object Columns):\n{df.describe(include='object').T}")

    print("\nMissing Values (Count and Percentage):")
    missing_values = df.isnull().sum()
    missing_percent = (missing_values / len(df)) * 100
    missing_df = pd.DataFrame({'Count': missing_values, 'Percentage': missing_percent})
    missing_df = missing_df[missing_df['Count'] > 0].sort_values(by='Percentage', ascending=False)
    
    sanitized_df_name = df_name.replace('.csv', '').replace(' ', '_') # Sanitize for filename

    if not missing_df.empty:
        print(missing_df)
        # Visualize missingness patterns
        fig_matrix = msno.matrix(df[missing_df.index], figsize=(12, 6), sparkline=False)
        plt.title(f'Missing Value Matrix for {df_name} (Columns with NaNs)', fontsize=16)
        plt.savefig(os.path.join(PLOTS_DIR, f'missing_matrix_{sanitized_df_name}.png'), bbox_inches='tight')
        plt.show()
        plt.close(fig_matrix.figure)


        if df.shape[1] <= 60: # Only plot bar if not too many columns
            fig_bar = msno.bar(df[missing_df.index], figsize=(12, 6), log=False)
            plt.title(f'Missing Value Bar Chart for {df_name} (Columns with NaNs)', fontsize=16)
            plt.savefig(os.path.join(PLOTS_DIR, f'missing_bar_{sanitized_df_name}.png'), bbox_inches='tight')
            plt.show()
            plt.close(fig_bar.figure)
    else:
        print("No missing values found.")

    entirely_empty_cols = df.columns[df.isnull().all()].tolist()
    if entirely_empty_cols:
        print(f"\nWARNING: The following columns are entirely empty (all NaNs): {entirely_empty_cols}")
    else:
        print("\nNo entirely empty columns found.")
    
    single_value_cols = [col for col in df.columns if df[col].nunique(dropna=True) == 1]
    if single_value_cols:
        print(f"\nWARNING: The following columns have only one unique value (excluding NaNs):")
        for col in single_value_cols:
            print(f"  - {col}: {df[col].dropna().unique()}")
    else:
        print("\nNo columns with only a single unique value found (excluding NaNs).")

    print(f"--- End Inspection for {df_name} ---")
    return missing_df

# --- Load Main DataFrames ---
print("Loading main dataframes for EDA...")
train_df = pd.read_csv(BASE_PATH + 'Train.csv')
test_df = pd.read_csv(BASE_PATH + 'Test.csv')
gap_train_df = pd.read_csv(BASE_PATH + 'Gap_Train.csv')
gap_test_df = pd.read_csv(BASE_PATH + 'Gap_Test.csv')
sample_submission = pd.read_csv(BASE_PATH + 'SampleSubmission.csv')
try:
    variable_definitions = pd.read_csv(BASE_PATH + 'VariableDefinitions.csv')
    print("\n--- Variable Definitions ---")
    print(variable_definitions)
except FileNotFoundError:
    print("\nVariableDefinitions.csv not found. Skipping.")


# --- EDA for Train.csv ---
print("\n================ EDA: Train.csv ================")
missing_train_info = inspect_dataframe(train_df, "Train.csv")

print("\n--- Target Variable Analysis (Train.csv) ---")
plt.figure(figsize=(18, 12))
for i, col in enumerate(TARGET_COLUMNS):
    plt.subplot(4, 3, i + 1)
    sns.histplot(train_df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'train_target_distributions.png'), bbox_inches='tight')
plt.show()
plt.close()

plt.figure(figsize=(10, 8))
sns.heatmap(train_df[TARGET_COLUMNS].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Target Nutrients')
plt.savefig(os.path.join(PLOTS_DIR, 'train_target_correlation_heatmap.png'), bbox_inches='tight')
plt.show()
plt.close()

print("\n--- Key Feature Analysis (Train.csv) ---")
key_features_train = ['pH', 'x', 'y', 'alb', 'bio1', 'bio12', 'cec20', 'soc20', 'BulkDensity', 'site'] \
                     + [col for col in train_df.columns if 'ls' in col or 'mb' in col or 'tim' in col or 'slope' in col]
key_features_train = [col for col in key_features_train if col in train_df.columns and col not in TARGET_COLUMNS]

numeric_key_features_train = train_df[key_features_train].select_dtypes(include=np.number).columns
if not numeric_key_features_train.empty:
    n_plots_numeric_key = min(20, len(numeric_key_features_train))
    n_rows_numeric_key = (n_plots_numeric_key + 3) // 4 # 4 plots per row
    
    plt.figure(figsize=(20, 5 * n_rows_numeric_key))
    for i, col in enumerate(numeric_key_features_train[:n_plots_numeric_key]):
        plt.subplot(n_rows_numeric_key, 4, i + 1)
        sns.histplot(train_df[col].dropna(), kde=True, bins=30) # Added dropna()
        plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_DIR, 'train_key_feature_distributions.png'), bbox_inches='tight')
    plt.show()
    plt.close()

    plt.figure(figsize=(20, 5 * n_rows_numeric_key))
    for i, col in enumerate(numeric_key_features_train[:n_plots_numeric_key]):
        plt.subplot(n_rows_numeric_key, 4, i + 1)
        sns.boxplot(y=train_df[col].dropna()) # Added dropna()
        plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_DIR, 'train_key_feature_boxplots.png'), bbox_inches='tight')
    plt.show()
    plt.close()
    print("Inspect boxplots above for potential outliers. IQR method can be used for programmatic detection.")
else:
    print("No numeric key features found in train_df for distribution plots.")

if 'site' in train_df.columns:
    plt.figure(figsize=(12, 8)) # Adjusted for better y-axis label visibility
    sns.countplot(y=train_df['site'], order=train_df['site'].value_counts().iloc[:20].index)
    plt.title('Distribution of Site (Top 20) in Train.csv')
    plt.xlabel('Count')
    plt.ylabel('Site ID')
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_DIR, 'train_site_distribution.png'), bbox_inches='tight')
    plt.show()
    plt.close()
    print(f"Number of unique sites in train: {train_df['site'].nunique()}")

if 'BulkDensity' in train_df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(train_df['BulkDensity'].dropna(), kde=True, bins=20)
    plt.title('Distribution of BulkDensity in Train.csv')
    plt.savefig(os.path.join(PLOTS_DIR, 'train_bulkdensity_distribution.png'), bbox_inches='tight')
    plt.show()
    plt.close()
    print(f"BulkDensity stats in train:\n{train_df['BulkDensity'].describe()}")

# --- EDA for Test.csv ---
print("\n================ EDA: Test.csv ================")
missing_test_info = inspect_dataframe(test_df, "Test.csv")

print("\n--- Comparing Key Feature Distributions (Train vs. Test) ---")
key_features_test = [col for col in key_features_train if col in test_df.columns and col != 'site']
numeric_key_features_test = test_df[key_features_test].select_dtypes(include=np.number).columns

if not numeric_key_features_test.empty:
    n_plots_compare = min(20, len(numeric_key_features_test))
    n_rows_compare = (n_plots_compare + 3) // 4

    plt.figure(figsize=(20, 5 * n_rows_compare))
    for i, col in enumerate(numeric_key_features_test[:n_plots_compare]):
        plt.subplot(n_rows_compare, 4, i + 1)
        sns.histplot(train_df[col].dropna(), kde=True, color='blue', label='Train', stat="density", common_norm=False)
        sns.histplot(test_df[col].dropna(), kde=True, color='red', label='Test', stat="density", common_norm=False)
        plt.title(f'Train vs. Test: {col}')
        plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(PLOTS_DIR, 'train_vs_test_key_feature_distributions.png'), bbox_inches='tight')
    plt.show()
    plt.close()
else:
    print("No common numeric key features found for train vs. test distribution plots.")

if 'site' in test_df.columns:
    print(f"Number of unique sites in test: {test_df['site'].nunique()}")
    common_sites = len(set(train_df['site'].unique()) & set(test_df['site'].unique()))
    print(f"Number of common sites between train and test: {common_sites}")
    new_sites_in_test = len(set(test_df['site'].unique()) - set(train_df['site'].unique()))
    print(f"Number of sites in test but not in train: {new_sites_in_test}")

if 'BulkDensity' in test_df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(test_df['BulkDensity'].dropna(), kde=True, bins=20)
    plt.title('Distribution of BulkDensity in Test.csv')
    plt.savefig(os.path.join(PLOTS_DIR, 'test_bulkdensity_distribution.png'), bbox_inches='tight')
    plt.show()
    plt.close()
    print(f"BulkDensity stats in test:\n{test_df['BulkDensity'].describe()}")


# --- EDA for Gap_Train.csv ---
print("\n================ EDA: Gap_Train.csv ================")
inspect_dataframe(gap_train_df, "Gap_Train.csv")
if 'Nutrient' in gap_train_df.columns and 'Required' in gap_train_df.columns:
    print("\nRequired values per Nutrient (Gap_Train.csv):")
    required_stats_train = gap_train_df.groupby('Nutrient')['Required'].agg(['mean', 'median', 'min', 'max', 'nunique']).sort_values(by='nunique', ascending=False)
    print(required_stats_train)
    # Optional: Plot 'Required' values if diverse enough
    # plt.figure(figsize=(10,6))
    # sns.boxplot(x='Nutrient', y='Required', data=gap_train_df)
    # plt.title('Required Nutrients in Gap_Train.csv')
    # plt.savefig(os.path.join(PLOTS_DIR, 'gap_train_required_boxplot.png'), bbox_inches='tight')
    # plt.show()
    # plt.close()


# --- EDA for Gap_Test.csv ---
print("\n================ EDA: Gap_Test.csv ================")
inspect_dataframe(gap_test_df, "Gap_Test.csv")
if 'Nutrient' in gap_test_df.columns and 'Required' in gap_test_df.columns:
    print("\nRequired values per Nutrient (Gap_Test.csv):")
    required_stats_test = gap_test_df.groupby('Nutrient')['Required'].agg(['mean', 'median', 'min', 'max', 'nunique']).sort_values(by='nunique', ascending=False)
    print(required_stats_test)

# --- EDA for SampleSubmission.csv ---
print("\n================ EDA: SampleSubmission.csv ================")
inspect_dataframe(sample_submission, "SampleSubmission.csv")
if 'ID' in sample_submission.columns:
    try:
        sample_submission[['PID_sub', 'Nutrient_sub']] = sample_submission['ID'].str.rsplit('_', n=1, expand=True)
        print("\nNutrient counts in SampleSubmission:")
        print(sample_submission['Nutrient_sub'].value_counts())
        plt.figure(figsize=(10, 6))
        sns.countplot(y=sample_submission['Nutrient_sub'], order=sample_submission['Nutrient_sub'].value_counts().index)
        plt.title('Nutrient Distribution in SampleSubmission.csv')
        plt.xlabel('Count')
        plt.ylabel('Nutrient')
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_DIR, 'sample_submission_nutrient_counts.png'), bbox_inches='tight')
        plt.show()
        plt.close()
    except Exception as e:
        print(f"Could not parse ID in SampleSubmission: {e}")


# --- EDA for Satellite Datasets ---
print("\n================ EDA: Satellite Datasets ================")
all_pids_main = pd.concat([train_df['PID'], test_df['PID']]).unique()
print(f"Total unique PIDs in Train+Test: {len(all_pids_main)}")

for file_name, prefix in SATELLITE_FILES_INFO.items():
    print(f"\n--- Processing Satellite File: {file_name} ---")
    sanitized_file_name = file_name.replace('.csv', '').replace(' ', '_')
    try:
        df_sat = pd.read_csv(BASE_PATH + file_name)
        inspect_dataframe(df_sat, file_name) # This already saves missingno plots

        if 'PID' in df_sat.columns:
            unique_pids_sat = df_sat['PID'].nunique()
            print(f"Unique PIDs in {file_name}: {unique_pids_sat}")
            pids_in_main_not_in_sat = len(set(all_pids_main) - set(df_sat['PID'].unique()))
            pids_in_sat_not_in_main = len(set(df_sat['PID'].unique()) - set(all_pids_main))
            print(f"PIDs in Train/Test but NOT in {file_name}: {pids_in_main_not_in_sat}")
            print(f"PIDs in {file_name} but NOT in Train/Test: {pids_in_sat_not_in_main}")

            obs_per_pid = df_sat.groupby('PID').size()
            plt.figure(figsize=(10, 5))
            sns.histplot(obs_per_pid, bins=min(50, obs_per_pid.nunique())) # Adjust bins
            plt.title(f'Observations per PID in {file_name}')
            plt.xlabel('Number of Observations')
            plt.ylabel('Count of PIDs')
            plt.savefig(os.path.join(PLOTS_DIR, f'sat_{sanitized_file_name}_obs_per_pid.png'), bbox_inches='tight')
            plt.show()
            plt.close()
            print(f"Stats for observations per PID in {file_name}:\n{obs_per_pid.describe()}")
        else:
            print(f"WARNING: 'PID' column not found in {file_name}.")

        date_cols = [col for col in df_sat.columns if any(sub in col.lower() for sub in ['date', 'time', 'day', 'month', 'year', '.geo'])]
        if date_cols:
            print(f"Potential date/time related columns: {date_cols}")
        else:
            print("No obvious date/time columns found by name.")

        numeric_sensor_cols = df_sat.select_dtypes(include=np.number).columns
        numeric_sensor_cols = [col for col in numeric_sensor_cols if col not in ['PID'] and not any(sub in col.lower() for sub in date_cols)]

        if len(numeric_sensor_cols) > 0:
            n_plots_sat_sensor = min(5, len(numeric_sensor_cols))
            print(f"Plotting distributions for a sample of numeric sensor columns in {file_name} (max {n_plots_sat_sensor}):")
            
            # Determine number of rows for subplot based on n_plots_sat_sensor
            if n_plots_sat_sensor == 1: n_rows_sat_sensor = 1
            elif n_plots_sat_sensor <= 3: n_rows_sat_sensor = 1
            else: n_rows_sat_sensor = (n_plots_sat_sensor + 2) // 3 # Aim for 2 or 3 plots per row

            n_cols_subplot = (n_plots_sat_sensor + n_rows_sat_sensor -1) // n_rows_sat_sensor


            plt.figure(figsize=(5 * n_cols_subplot, 4 * n_rows_sat_sensor)) # Adjusted figsize
            for i, sensor_col in enumerate(numeric_sensor_cols[:n_plots_sat_sensor]):
                plt.subplot(n_rows_sat_sensor, n_cols_subplot, i + 1)
                sample_size = min(len(df_sat), 50000)
                sns.histplot(df_sat[sensor_col].sample(sample_size, random_state=SEED).dropna(), kde=False, bins=30)
                plt.title(sensor_col, fontsize=10) # Smaller title for tight layout
            plt.tight_layout()
            plt.savefig(os.path.join(PLOTS_DIR, f'sat_{sanitized_file_name}_sensor_distributions.png'), bbox_inches='tight')
            plt.show()
            plt.close()
        else:
            print(f"No numeric sensor columns found in {file_name} for distribution plots.")

        del df_sat
        gc.collect()

    except FileNotFoundError:
        print(f"ERROR: {file_name} not found. Skipping.")
    except Exception as e:
        print(f"An error occurred while processing {file_name}: {e}")


print("\n================ EDA Complete ================")
print(f"All plots have been saved to the '{PLOTS_DIR}' directory.")
print("Key things to note from EDA for next steps (Feature Engineering & Preprocessing):")
print("1. Missing values: Decide on imputation strategies.")
print("2. Outliers: Consider their impact. Transformation or capping might be needed.")
print("3. Feature Scaling: Important for some models.")
print("4. Categorical Features: 'site' needs encoding.")
print("5. Satellite Data Aggregation: Decide on methods (mean, median, std, trends, etc.).")
print("6. Column Names: Ensure consistency.")
print("7. Distribution Drift: Note differences between train and test.")
print("8. Empty/Single-Value Columns: Can usually be dropped.")
```
